# Q6. Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
    1. Brand
    2. Product Description
    3. Price
    4. Discount %

The attributes which you have to scrape is ticked marked in the below image.

To scrape the data you have to go through following steps:
    1. Go to flipkart webpage by url https://www.flipkart.com/
    2. Enter “sunglasses” in the search field where “search for products, brands and more” is written and click the search icon
    3. after that you will reach to a webpage having a lot of sunglasses. From this page you can scrap the required data as usual.
    4. after scraping data from the first page, go to the “Next” Button at the bottom of the page , then click on it.
    5. Now scrape data from this page as usual
    6. repeat this until you get data for 100 sunglasses.

```python
#Calling the driver to load the webpage
driver=webdriver.Chrome(r"C://Users//Aditya//Downloads//chromedriver.exe")
driver.get("https://www.flipkart.com/")
driver.maximize_window()

#Closing the Login Pop-up
time.sleep(3)
Login_pop_up = driver.find_element_by_xpath("//div[@class='_2QfC02']/button[@class = '_2KpZ6l _2doB4z']")
Login_pop_up.click()

#Seaching for Sunglasses
product_search = driver.find_element_by_xpath("//div[@class='_3OO5Xc']//input[@type = 'text']")
product_search.send_keys("sunglasses")

search_button = driver.find_element_by_xpath("//div[@class = 'col-12-12 _2oO9oE']//button[@class = 'L0Z3Pu']")
search_button.click()
```

```python
#Creating Lists for attributes
Brand = []
Product= []
Price = []
Discount =[]

#Coding for 100 Sunglasses Data Scraping
#The required data is present in 3 webpages. As each webpage contains 40 sunglasses.

while True:
    
    time.sleep(3)
        
    for i in driver.find_elements_by_xpath("//div[@class= '_2B099V']//div[@class = '_2WkVRV']"):
        Brand.append(i.text)
        if len(Brand) == 100:
            break
    
    for i in driver.find_elements_by_xpath("//div[@class= '_2B099V']//a"):
        Product.append(i.get_attribute('title'))
        if len(Product) == 200:
            break
    
    for i in driver.find_elements_by_xpath("//div[@class= '_25b18c']//div[@class = '_30jeq3']"):
        Price.append(i.text) 
        if len(Price) == 100:
            break
    
    for i in driver.find_elements_by_xpath("//div[@class= '_3Ay6Sb']/span"):
        Discount.append(i.text)
        if len(Discount) == 100:
            break
    
    if len(Brand) == 100:
        break
            
    #For loop for next page to scrap data  
    for j in driver.find_elements_by_xpath("//nav[@class ='yFHi8N']/a[@class = '_1LKTO3']"):
        next_page=[j.get_attribute("href")]
    
    time.sleep(3)
        
    for n in next_page:
        driver.get(n)
        
driver.close()
```

```python
print(len(Brand),len(Product),len(Price),len(Discount))
```

```python
#Shortening Product Description by cleaning unwanted info
Product_Description = []
for i in range(len(Product)):
    if i == 0 or  i/2 == i//2:
        Product_Description.append(Product[i])
```

```python
print(len(Brand),len(Product_Description),len(Price),len(Discount))
```

```python
#Creating DataFrame.
Flipkart_100_Sunglasses = pd.DataFrame()
Flipkart_100_Sunglasses["Brand"]=Brand
Flipkart_100_Sunglasses["Product Description"]=Product_Description
Flipkart_100_Sunglasses["Price"]=Price
Flipkart_100_Sunglasses["Discount"]=Discount

pd.set_option("display.max_rows", None,"display.max_colwidth",None)
Flipkart_100_Sunglasses
```
