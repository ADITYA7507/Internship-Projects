# Q1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
    
You need to find following details:
    
A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
#importing the required library
from selenium import webdriver
import pandas as pd
import time
```

```python
#Calling the Web Driver and loading the target webpage
driver = webdriver.Chrome(r"C://Users//Aditya//Downloads//chromedriver.exe")
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
driver.maximize_window() #Maximize the Window
```

```python
#Creating the list of Top 30 Most Videos with Details from the webpage
details=[]

for i in driver.find_elements_by_xpath("//table [@class = 'wikitable sortable jquery-tablesorter']/tbody/tr/td"):
        details.append(i.text)
        
details=details[:180] #As there is 30 videos list and 6 attributes. Storing the required details

driver.close()

Rank = [] #list for storing the Rank of the video.
Name = [] #list for storing the Name of the video.
Artist = [] #list for storing the Artist of the video.
Views = [] #list for storing the Views of the video.
Upload_date = [] #list for storing the Uploaded date of the video.

i= 0 

while i < len(details):
    Rank.append(details[i])
    i+=1
    Name.append(details[i])
    i+=1
    Artist.append(details[i])
    i+=1
    Views.append(details[i])
    i+=1
    Upload_date.append(details[i])
    i+=2
```

```python
#Creating DataFrame
Most_viewed_Videos = pd.DataFrame({"Rank" : Rank,
                          "Name" : Name, 
                          "Artist" : Artist,
                          "Upload date" : Upload_date,
                          "Views(in Million)" : Views})

Most_viewed_Videos
```

# Q2. Scrape the details team India’s international fixtures from bcci.tv.

Url = https://www.bcci.tv/
    
You need to find following details:
    
A) Match title (i.e. 1st ODI)

B) Series

C) Place

D) Date

E) Time

Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
#Calling the Web Driver and loading the target webpage
driver = webdriver.Chrome(r"C://Users//Aditya//Downloads//chromedriver.exe")
driver.get("https://www.bcci.tv/")
driver.maximize_window() #Maximize the Window

btn = driver.find_element_by_xpath("//div[@class = 'navigation__drop-down drop-down drop-down--reveal-on-hover']/div/ul/li/a")
driver.get(btn.get_attribute("href"))
```

```python
Match_title = [] #Creating a List for storing Match titles like Test, ODI or T20I
Series = [] #Creating a List for storing the name of the Series 
Place = [] #Creating a List for storing Venue of the Match.
Date = [] #Creating a List for storing the date of the Match.
Time = [] #Creating a List for storing the time of the Match.
datetime = [] #Creating a List of Date and Time got from scrapping
    
#Scraping all details for each list.
for i in driver.find_elements_by_xpath("//div [@class = 'fixture__format-strip']/ span [@class = 'u-unskewed-text fixture__format']"):
    Match_title.append(i.text)
       
for i in driver.find_elements_by_xpath("//div [@class = 'fixture__format-strip']/ span [@class = 'u-unskewed-text fixture__tournament-label u-truncated']"):
    Series.append(i.text)
    
for i in driver.find_elements_by_xpath("//div [@class = 'fixture__description u-unskewed-text']/p/span"):
    Place.append(i.text)
    
for i in driver.find_elements_by_xpath("//div [@class = 'fixture__datetime desktop-only']"):
    datetime.append(i.text.replace('\n',' '))
    
    
driver.close()   
    
#Spliting the Date and Time. 
Date = [i.split(' ',3)[:3]for i in datetime]
Date = [' '.join(i) for i in Date]
Time = [i.split(' ',3)[-1]for i in datetime]
```

```python
#Creating the DataFrame
International_Fixtures = pd.DataFrame({'Match title' : Match_title , 
                                            'Series' : Series , 
                                             'Place' : Place, 
                                              'Date' : Date ,
                                              'Time' : Time})

International_Fixtures
```

# Q5. Scrape the details of trending repositories on Github.com.

Url = https://github.com/
    
You have to find the following details:
    
A) Repository title

B) Repository description

C) Contributors count

D) Language used

Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
#Calling the Web Driver and loading the URL and target webpage
driver = webdriver.Chrome(r"C://Users//Aditya//Downloads//chromedriver.exe")
driver.get("https://github.com/")
driver.maximize_window() #Maximize the Window

button = driver.find_element_by_xpath("//ul [@class= 'list-style-none mb-3']/li [@class = 'edge-item-fix']/a [@data-ga-click = '(Logged out) Header, go to Trending']")
driver.get(button.get_attribute('href'))

time.sleep(3)
    
#Storing all the links of repositories    
href = [] 
for i in driver.find_elements_by_xpath("//h1 [@class = 'h3 lh-condensed']/a"):
    href.append(i.get_attribute('href'))
        
driver.close()
    
#Creating lists for required attributes for scrapping
Repository_title = [] 
Repository_description = [] 
Contributors_count = []
Language_used = [] 
    
#Opening each link saved in href and scrap the details
for i in href:
    driver = webdriver.Chrome(r"C://Users//Aditya//Downloads//chromedriver.exe")
    driver.get(i)
    driver.maximize_window()
    
    time.sleep(2) 
        
    #Scraping the Repository_title
    tit = driver.find_element_by_xpath("//h1 [@class = ' d-flex flex-wrap flex-items-center break-word f3 text-normal']")
    Repository_title.append(tit.text.replace('\n', ' '))
        
    #Scraping the Repository_description.
    try:
        des = driver.find_element_by_xpath("//div [@class = 'BorderGrid-cell']/p [@class = 'f4 mt-3']")
        Repository_description.append(des.text)
    except:
        Repository_description.append("-")
        
    #Scraping the Contributors_count.
    count = [] 
    try:
        for i in driver.find_elements_by_xpath("//h2 [@class = 'h4 mb-3']//span [@class = 'Counter ']"):
            count.append(i.get_attribute('title'))
            Contributors_count.append(count[-1])
    except:
        Contributors_count.append('1')
                
    #Scraping the Language_used.
    lang = []
    for i in driver.find_elements_by_xpath("//ul [@class = 'list-style-none']/li [@class = 'd-inline']//span"):
        lang.append(i.text)
    if (" ".join(lang)) == '':
        Language_used.append('-')
    else:
        Language_used.append(" ".join(lang))
        
    driver.close()
```

```python
print(len(Repository_title),len(Repository_description),len(Contributors_count),len(Language_used))
```

```python
Contributors_count=Contributors_count[:25]
```

```python
#Creating the DataFrame
Repository = pd.DataFrame({"Repository title" : Repository_title,
                          "Repository description" : Repository_description,
                          "Contributors count" : Contributors_count,
                          "Language used" : Language_used})
```

# Q7. Scrape the details of Data science recruiters from naukri.com.

Url = https://www.naukri.com/
    
You have to find the following details:
    
A) Name

B) Designation

C) Company

D) Skills they hire for

E) Location

Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code

```python
#Calling the Web Driver and loading the URL and target webpage
driver = webdriver.Chrome(r"C://Users//Aditya//Downloads//chromedriver.exe")
driver.get("https://www.naukri.com/")
driver.maximize_window() #Maximize the Window

#Landing on the Recruiters page
Recruiters = driver.find_element_by_xpath("//li/a [@title = 'Search Recruiters']") 
href= Recruiters.get_attribute('href')
driver.get(href)
    
#Searching the input key and sending the key as Data Science.
search = driver.find_element_by_xpath("//div[@class= 'inpWrap']/ input [@class = 'sugInp']")
search.send_keys("Data science")
    
#Searching the search Button and Clicking it.
search_btn = driver.find_element_by_xpath("//button[@class = 'fl qsbSrch blueBtn']")
search_btn.click()
    
#Creating a list and storing details of name and Company of Recuriters.
details = [i.text for i in driver.find_elements_by_xpath("//p [@class ='highlightable']//a[@class = 'ellipsis']")]
    
#Creating a list and storing details of href of Recuriters.
href = [i.get_attribute('href') for i in driver.find_elements_by_xpath("//p [@class ='highlightable']//a[@class = 'ellipsis']")]
    
#Creating a list and storing details of Designation of Recuriters.
Designation = [i.text for i in driver.find_elements_by_xpath("//p [@class ='highlightable']//span[@ class= 'ellipsis clr']")]
    
#Creating a list and storing details of skills of Recuriters.
skills = [i.text for i in driver.find_elements_by_xpath("//div [@class ='recInfo']//div[@ class= 'hireSec highlightable']")]

#Creating list for the attributes
Name = []
company = []
hreff = []
Location = []
 
#Extracting data to required attributes from details and href list
j = 0
while j < len(details):
    Name.append(details[j])
    hreff.append(href[j])
    j+=1
    company.append(details[j])
    j+=1
    
#Storing locations of the recruiters in the Location list
for i in hreff:
    driver.get(i)    
    loc = [i.text for i in driver.find_elements_by_xpath("//div [@class ='oh']/a[@ target= '_blank']")]
    loc = loc[-1]
    if loc == 'Others':
        Location.append('-')
    elif loc:
        Location.append(loc)
    else:
        Location.append('-')

driver.quit()
```

```python
#Creating a DataFrame
Recruiters = pd.DataFrame({"Name" : Name,
                          "Company" : company,
                          "Designation" : Designation,
                          "Location" : Location,
                          "skills": skills})

Recruiters
```
