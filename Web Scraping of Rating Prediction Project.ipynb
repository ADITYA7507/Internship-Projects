```python
# Importing Libraries
import selenium
import pandas as pd
import time

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
#Defining a Function for Amazon
def amazon(user_ip):
    #Search bar page accesing steps
    driver = webdriver.Chrome(r"F://chromedriver.exe")
    driver.get('https://www.amazon.in/')
    search_bar = driver.find_element_by_id("twotabsearchtextbox")   
    search_bar.clear()                                              
    search_bar.send_keys(user_ip)                                   
    search_button = driver.find_element_by_xpath('//div[@class="nav-search-submit nav-sprite"]/span/input') 
    search_button.click()
    
    HREF=[]  #Product link list declaration
    n=0
    
    #Scraping product links
    while True:
        
        for i in driver.find_elements_by_xpath("//h2/a"):
            HREF.append(i.get_attribute('href'))
            if len(HREF)==50:
                break
    
        if len(HREF)==50:
            break
        
        for j in driver.find_elements_by_xpath("//li[@class='a-last']/a"):
            next_page=[j.get_attribute("href")]
        
        for n in next_page:
            driver.get(n)

    driver.close()
    
    
    Ratings=[]
    Reviews=[]
    
    #Scraping Ratings and Reviews
    for i in HREF:
        driver = webdriver.Chrome(r"F://chromedriver.exe")
        driver.get(i+'#customerReviews')
        driver.maximize_window()
    
        time.sleep(2)
        
        STAR=[]
        
        for s in driver.find_elements_by_xpath("//td[@class='aok-nowrap']/span[@class='a-size-base']/a"):
            STAR.append(s.get_attribute('href'))
    
        for q in STAR:
            driver.get(q)
        
            try:
                for rh in driver.find_elements_by_xpath("//div[@class = 'a-row']/a[@class='a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold']"):
                    Reviews.append(rh.text)
            
            except:
                continue
        
            try:
                for rt in driver.find_elements_by_xpath("//div[@class = 'a-row']/a"):
                    Ratings.append(rt.get_attribute('title'))
            
            except:
                Ratings.append('-')
        
        driver.close()
        
    
    Ratings=Ratings[0:len(Ratings):2]

    #Creating DataFrame
    Rating_Review=pd.DataFrame({"Rating":Ratings,
                               "Review":Reviews})
    
    
    return Rating_Review
```

## Now,we can call amazon function for each product rating.

For Loop is not used here because of long time taking for a single iteration as 100 href links are applied for each product and each link has 10 reviews.

And any disturbance in connection can be led to abort the for loop and have to start again from the first product.

# Web Scraping using Flipkart

```python
def flipkart(user_ip):
    
    #Calling the driver to load the webpage
    driver=webdriver.Chrome(r"F://chromedriver.exe")
    driver.get("https://www.flipkart.com/")
    driver.maximize_window()

    #Closing the Login Pop-up
    time.sleep(1)
    Login_pop_up = driver.find_element_by_xpath("//div[@class='_2QfC02']/button[@class = '_2KpZ6l _2doB4z']")
    Login_pop_up.click()

    product_search = driver.find_element_by_xpath("//div[@class='_3OO5Xc']//input[@type = 'text']")
    product_search.send_keys(user_ip)

    search_button = driver.find_element_by_xpath("//div[@class = 'col-12-12 _2oO9oE']//button[@class = 'L0Z3Pu']")
    search_button.click()
    
    href=[]
    Review_details=[]
    Rating=[]
    Review=[]
    next_page=[]

    #Scraping Product links
    while True:
    
        for i in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
            href.append(i.get_attribute('href'))
            if len(href) == 50:
                break
    
        if len(href) == 50:
            break
        
        for j in driver.find_elements_by_xpath("//a[@class = '_1LKTO3']"):
            next_page.append(j.get_attribute("href"))
        
        for n in next_page:
            driver.get(n)
        
    driver.close()
    
    #Scraping Ratings and Reviews
    for i in href:
        driver=webdriver.Chrome(r"F://chromedriver.exe")
        driver.get(i)
        
        try:
            All_Reviews=driver.find_elements_by_xpath("//div[@class='col JOpGWq']/a")
            AR=All_Reviews[0].get_attribute('href')
            driver.get(AR)
            driver.maximize_window()
        
        except:
            driver.close()
            continue

            
        try:
            for rev in driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']"):
                Review_details.append(rev.text)

        except:
            continue
            
        
        NRV=driver.find_elements_by_xpath("//select[@class='_1EDlbo tVKh2S']//option[@value='NEGATIVE_FIRST']")
        NRV[0].click()
        
        try:
            for nr in driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']"):
                Review_details.append(nr.text)
                
        except:
            continue
    
        driver.close()
    
    for i in range(len(Review_details)):
        Review_details[i]=Review_details[i].split('\n')
            
        
    for i in range(len(Review_details)):
        Rating.append(Review_details[i][0])
        Review.append(Review_details[i][1])
        
    #Creating DataFrame        
    Rating_Review=pd.DataFrame({"Rating":Rating,
                               "Review":Review})
    
    
    return Rating_Review
```

```python
Laptop_Rating_Review_Flipkart=flipkart("laptops")
Laptop_Rating_Review_Flipkart.to_csv("Laptop_Rating_Review_Flipkart.csv")
Laptop_Rating_Review_Flipkart
```
